## Step 1: Read the image in RGB format

```python
import cv2
import matplotlib.pyplot as plt

# Read the image in RGB format
original_image = cv2.imread("cat_image2.jpeg")

# Display the original image
original_image = cv2.cvtColor(original_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(original_image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

## Step 2: Convert the image to grayscale

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.figure(figsize=(8,8))
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

## Step 3: Invert the grayscale image (negative image)

```python
# Invert the grayscale image (negative image)
inverted_image = 255 - gray_image

# Display the inverted image
plt.figure(figsize=(8,8))
plt.imshow(inverted_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

## Step 4: Blur the inverted image

```python
# Blur the inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)

# Display the blurred inverted image
plt.figure(figsize=(8,8))
plt.imshow(blurred_image,cmap="gray")
plt.axis("off")
plt.title("Blurred Image")
plt.show()
```

## Step 5: Create the pencil sketch by dividing the grayscale image by the inverted blurry image

```python
# Create the pencil sketch by dividing the grayscale image by the inverted blurry image
pencil_sketch_image = cv2.divide(gray_image, 255 - blurred_image, scale=255)

# Display the pencil sketch image
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch_image, cmap="gray")
plt.axis("off")
plt.title("Final Pencil Sketch Image")
plt.show()
```

## Let's compare the original image with the pencil sketch side by side.

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(original_image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(pencil_sketch_image,cmap="gray")
plt.axis("off")
plt.title("Final Pencil Sketch Image")
plt.show()
```
